```python

from transformers import pipeline

pipe = pipeline("text-classification", 
                model="j-hartmann/emotion-english-distilroberta-base",
                top_k=None,
                device=0,
                truncation=True,                    
                max_length=512,                     
                truncation_strategy='longest_first' 
)
```

```python
def calculate_max_scores(description):
    sentences = [s.strip() for s in description.split('.') if s.strip()]
    
    sentence_emotions = []
    for sentence in sentences:
        scores = pipe(sentence)
        max_emotion = max(scores[0], key=lambda x: x["score"])
        sentence_emotions.append([max_emotion['label'],max_emotion['score']])
    return max(sentence_emotions)
def calculate_max_scores(description):
    if not description or not isinstance(description, str) or not description.strip():
        return ['neutral', 0.0]  
    
    sentences = [s.strip() for s in description.split('.') if s.strip()]
    
    if not sentences:
        return ['neutral', 0.0]
    
    sentence_emotions = []
    for sentence in sentences:
        try:
            scores = pipe(sentence)
            
            # Check if scores list is empty
            if not scores or not scores[0]:
                continue
            
            max_emotion = max(scores[0], key=lambda x: x["score"])
            sentence_emotions.append([max_emotion['label'], max_emotion['score']])
        
        except Exception as e:
            print(f"Error processing sentence: {sentence}")
            print(f"Error details: {e}")
            continue
    
    if not sentence_emotions:
        return ['neutral', 0.0]
    
    return max(sentence_emotions, key=lambda x: x[1])
```

```python
res = calculate_max_scores(df["Description"][2])
res
```

```python
from tqdm import tqdm
tqdm.pandas()
df["Emotion"] = df["Description"].progress_apply(calculate_max_scores)
```

```python
df['sentiment'] = df['Emotion'].apply(lambda x: x[0] if x else 'neutral')
```
