```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
books["Genres"] = books["Genres"].replace("[]",np.nan)
```

```python
books.isna().sum()
```

```python

books["Num_Ratings"] = books["Num_Ratings"].str.replace(",", "").astype(int)
```

```python
books_sorted_popularity = books.sort_values(by=["Num_Ratings"],ascending=False)
```

```python
books_sorted_popularity
```

```python
books_sorted_popularity.columns
```

```python
def add_book_identifiers(df):
    # Create a copy to avoid modifying the original
    df_with_ids = df.copy()
    
    # Add a simple numeric ID starting from 1
    df_with_ids['BookID'] = range(1, len(df) + 1)
    
    # Create a formatted ID with 'BK' prefix and leading zeros
    # e.g., BK000001, BK000002, etc.
    df_with_ids['FormattedBookID'] = df_with_ids['BookID'].apply(
        lambda x: f'BK{str(x).zfill(6)}'
    )
    
    # Reorder columns to put IDs first
    cols = ['BookID', 'FormattedBookID'] + [col for col in df.columns if col != 'Unnamed: 0']
    df_with_ids = df_with_ids[cols]
    
    # Drop the 'Unnamed: 0' column if it exists
    if 'Unnamed: 0' in df_with_ids.columns:
        df_with_ids = df_with_ids.drop('Unnamed: 0', axis=1)
    
    return df_with_ids
books_with_Identifiers = add_book_identifiers(books_sorted_popularity)
```

```python
books_with_Identifiers
```

```python
books["id_desc"] = books["FormattedBookID"] +" "+ books["Description"]
books.to_csv("books_data.csv",index=False)
```
